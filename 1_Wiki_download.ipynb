```python
import wikipedia
import os

# Function to download and save Wikipedia page content for a given language
def download_and_save_page(page_title, lang, main_folder, sub_folder):
    # Set language
    wikipedia.set_lang(lang)

    # Get page object
    try:
        page = wikipedia.page(page_title)
    except wikipedia.exceptions.PageError:
        print(f"Page '{page_title}' not found in language '{lang}'.")
        return
    except wikipedia.exceptions.DisambiguationError as e:
        print(f"Disambiguation error for '{page_title}' in language '{lang}': {e.options}")
        return

    # Get page content
    content = page.content

    # Construct the path for the subfolder inside the main folder
    sub_folder_path = os.path.join(main_folder, sub_folder)

    # Ensure the subfolder exists
    os.makedirs(sub_folder_path, exist_ok=True)

    # Save to file
    filename = os.path.join(sub_folder_path, "{}_{}.txt".format(page_title, lang))
    print("Saving file to:", filename)  # Print the file path
    try:
        with open(filename, "w", encoding="utf-8") as file:
            file.write(content)
        print("Downloaded {} successfully. Saved at: {}".format(filename, os.path.abspath(filename)))
    except Exception as e:
        print("Failed to save file:", e)

# Dictionary of page titles and their translations
page_titles = {
    "Sostanza grigia": {"it": "Sostanza grigia",
              "en": "Grey matter",
              "es": "Sustancia gris",
              "de": "Graue Substanz",
              "fr": "Substance grise"}
}

# Languages to download
languages = ["it", "en", "fr", "es", "de"]

# Main folder name
main_folder = "/content/drive/MyDrive/Wiki download/Science_en_it_es_de_fr"  # Replace with your desired main folder path

# Iterate over page titles and languages to download and save page content
for title, translations in page_titles.items():
    for lang in languages:
        page_title = translations.get(lang)
        if page_title:
            # Create a subfolder based on the title
            sub_folder = title.replace(" ", "_")  # Replace spaces with underscores for folder name
            print("Downloading {} in {}...".format(title, lang))
            download_and_save_page(page_title, lang, main_folder, sub_folder)
```

```python
import wikipedia

# Function to download and save Wikipedia page content for a given language
def download_and_save_page(page_title, lang):
    # Set language
    wikipedia.set_lang(lang)

    # Get page object
    page = wikipedia.page(page_title)

    # Get page content
    content = page.content

    # Save to file
    filename = "{}_{}.txt".format(page_title, lang)
    with open(filename, "w", encoding="utf-8") as file:
        file.write(content)

    print("Downloaded {} successfully.".format(filename))

# Page title
page_title = "Meninges"
# Languages to download
languages = ["en"]

# Download and save page content for each language
for lang in languages:
    print("Downloading {}...".format(lang))
    download_and_save_page(page_title, lang)
```
